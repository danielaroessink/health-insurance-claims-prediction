# Data Cleaning and Feature engineering
## Overview
This notebook focuses on preparing the data for modeling by:
1. Cleaning the data (handling duplicates, inconsistencies, and outliers).
2. Encoding categorical variables.
3. Scaling and transforming features.
4. Creating and selecting relevant features.
5. Splitting the data into training, validation, and testing subsets.


```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Setting a style for plots
sns.set(style="whitegrid")

# Load the dataset
file_path = '../data/raw_data.csv'
df = pd.read_csv(file_path)

# Quick preview of the dataset
df.head()
```

```python
df.info()
```

## 1. Data Cleaning
During EDA we already established that there are no missing values in any of the variables so we can skip this step.

### 1.1. Check and remove duplicates

```python
# Check for duplicates
print(f"Number of duplicate rows: {df.duplicated().sum}")

# Remove duplicates if any
df = df.drop_duplicates()
print(f"Shape of dataset after removing duplicates: {df.shape}")
```

### 1.2. Fix Incosistent data

```python
# Example: Ensuring consistent case in a categorical column (Region)
df['region'] = df['region'].str.lower()

# Checking unique values in categorical columns for inconsistensies
for col in df.select_dtypes(include='object').columns:
    print(f"Unique values in {col}: {df[col].unique()}")
```

### 1.3. Outliers Handling
Using boxplots to identify outliers and decide how to handle them.

```python
# Boxplot for numerical columns
numerical_cols = df.select_dtypes(include=['float64','int64']).columns
for i, col in enumerate(numerical_cols):
    #plt.figure(figsize=(6,4))
    plt.subplot(2,2,i+1)
    sns.boxplot(df[col])
plt.title(f"Boxplot of {col}")
plt.show()
```

**Key Insights from Boxplot Analysis:**
- BMI shows several high outliers that may need transformation or capping
- Age distribution is relatively normal with no significant outliers to address
- Children is a discrete count variable (0-5) with right-skewed distribution
- Consider robust algorithms **less sensitive to outliers (Random Forest, XGBoost)**
- BMI likely needs normalization through **log transformation**
- Children variable may perform better with categorical encoding rather than numeric representation

**Treatment approaches:**
- Log transformation: log(BMI) to normalise the right sweked distribution
- Winsorisation: Cap values above 95th percentile rather than removing them
- Create binary flag: add a feature "is_high_BMI" to preserve the information

**Model-specific hnadling**:
- if using tree-based models (Random forest / XGBoost): keep original values as these models are robust to handle outliers
- If using linear models: use transformed version (log BMI)

**Validation step**:
- compare model performance with different outlier treatments
- ensure predictions for high BMI individuals remain clinically meaningful 

**DECISION AT THIS POINT**: I'm leaving outliers as they are as I would use a tree-based model which is robust enough to handle them.

## 2. Categorical Variable Encoding
**Goal:** Convert categorical variable to numerical formats for machine learning models.

```python
# Encoding Categorical columns
categorical_cols = df.select_dtypes(include='object').columns
print(f"Categorical columns: {categorical_cols}")
```

```python
# For binary variables (gender, smoker)
label_encoder = LabelEncoder()
df['sex_encoded'] = label_encoder.fit_transform(df['sex'])
df['smoker_encoded'] = label_encoder.fit_transform(df['smoker'])

# For region (nominal variable)
encoder = OneHotEncoder(sparse_output=False, drop='first') # drop first to avoid multicollinearity
region_encoded = encoder.fit_transform(df[['region']])
region_df = pd.DataFrame(region_encoded, columns=[f'region_{cat}' for cat in encoder.categories_[0][1:]])
```

Children is a discrete count variable (0-5) with a right skewed distribution. Since it is a count variable with a limited range, it might be better treated as a categorical vairable through one-hot encoding rather than scaling it numerically

```python
# For children (treating as categorical):
children_encoder = OneHotEncoder(sparse_output=False, drop='first')
children_encoded = children_encoder.fit_transform(df[['children']])
children_df = pd.DataFrame(children_encoded, columns=[f'child_{cat}' for cat in children_encoder.categories_[0][1:]])
```

```python
# Combine with original dataframe
df = pd.concat([df, region_df, children_df], axis=1)
df.head()
```

## 3. Feature scaling and Transformation
**Goal**: Scale and transform numerical features to ensure comparability and reduce skewness.

```python
from sklearn.preprocessing import StandardScaler

# For right-skewed continuous variables
df['log_bmi'] = np.log1p(df['bmi']) # log1p avoids issues with zeros

# Scaling numerical features
scaler = StandardScaler()
scale_cols = ['age', 'log_bmi'] # Standarisation after log transformation ensures all features are on comparable scales
scaled_features = scaler.fit_transform(df[scale_cols])
scaled_df = pd.DataFrame(scaled_features, columns=[col + '_scaled' for col in scale_cols])
```

```python
# Combine with original dataframe
df = pd.concat([df, scaled_df], axis=1)
df.head()
```

```python
# Create a copy of the dataframe with the scaled and encoded columns only
df_copy = df.copy().drop(columns=['age','sex','bmi','smoker','region','children','log_bmi'])
df_copy.head()
```

## 4. Save processed DataFrame

```python
# Save the preprocessed data to a CSV file
df_copy.to_csv('../data/processed_data.csv', index=False)
```
